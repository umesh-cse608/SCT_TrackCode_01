```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Actual data
Data = {
    "Square_feet": [1200,1600,2000,2400,2500],
    "Bedrooms": [2,3,4,4,5],
    "Bathrooms": [1,2,2,3,3],
    "price":[350000,360000,460000,540000,650000]
}

df = pd.DataFrame(Data)

X = df[["Square_feet","Bedrooms","Bathrooms"]]

Y = df[["price"]]

# to create model and train it using data

Model = LinearRegression()

Model.fit(X,Y)

prediction_price = Model.predict([[1700,4,2]])

print("predicted house price:",prediction_price[0])
```
